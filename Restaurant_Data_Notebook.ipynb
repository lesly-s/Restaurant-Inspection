```
import numpy as np
import pandas as pd
import sqlite3
```

```
df = pd.read_csv('/work/restaurant_modified_data.csv')
df
```

```
# feature engineering: removing columns 
df.drop(['Community Board', 'Council District', 'Census Tract','BIN', 'BBL', 'NTA'], axis = 1, inplace = True)
df
```

```
# changing grade values
grade_change_df = df.copy()
grade_change_df
```

```
# changing grade values based on score
grade_change_df.loc[
    (grade_change_df['GRADE'].isna()) & 
    (grade_change_df['INSPECTION DATE'] == grade_change_df['GRADE DATE']), 
    'GRADE'
] = 'Z'
grade_change_df.loc[grade_change_df['GRADE DATE'].isna(), 'GRADE'] = 'N'
grade_change_df.loc[grade_change_df['GRADE'] == 'P', 'GRADE'] = 'P'
grade_change_df.loc[(grade_change_df['GRADE'].isna()) & (grade_change_df['SCORE'] < 14), 'GRADE'] = 'A'
grade_change_df.loc[(grade_change_df['GRADE'].isna()) & (grade_change_df['SCORE'].between(14, 17)), 'GRADE'] = 'B'
grade_change_df.loc[(grade_change_df['GRADE'].isna()) & (grade_change_df['SCORE'] > 28), 'GRADE'] = 'C'

grade_change_df
```

```
#remove rows with inspection date = 1900 
grade_change_df = grade_change_df.loc[
    ~((grade_change_df['INSPECTION DATE'] == '1/1/00') & (grade_change_df['CRITICAL FLAG'] == 'Not Applicable'))
]
grade_change_df 
```

```
#remove duplicates
grade_change_df = grade_change_df.drop_duplicates()
grade_change_df
```

# Grade Range

```
grade_range_df = df.copy()
grade_range_df
```

```
#creating column Grade Range
grade_range_df.insert(15, 'GRADE RANGE', df['GRADE'])
grade_range_df['GRADE RANGE'] = grade_range_df['GRADE RANGE'].replace('A', '0-13')
grade_range_df['GRADE RANGE'] = grade_range_df['GRADE RANGE'].replace('B', '14-27')
grade_range_df['GRADE RANGE'] = grade_range_df['GRADE RANGE'].replace('C', '28+')
grade_range_df['GRADE RANGE'] = grade_range_df['GRADE RANGE'].replace('Z', 'grade pending')
grade_range_df
```

# Data for downloading

```
grade_change_df.to_csv('/work/down_modified_dataset.csv', index=False)
```
